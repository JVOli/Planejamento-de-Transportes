```python
file = r'Dados.xls'         #Ler o excel com os dados da pesquisa

df = pd.read_excel(file)    #Transformar os dados em um dataframe

df.head()                          #Mostra o dataframe criado
```

```python
residencias = {}                                                                    #Cria um dicionario para armazenar os Código do domicílio
temp = []                                                                           #Cria uma lista que sera usada para verificar se a zona ja apareceu antes
domicilios = {}                                                                     #Cria um dicionario para armazenar os Fator de expansão para domicílios
população = {}                                                                      #Cria um dicionario para armazenar as populações
automoveis = {}                                                                     #Cria um dicionario para armazenar os automoveis

for line in df.index:                                                               #Loop em todas as linhas do dataframe

    if df.loc[line, "Zona de Tráfego do domicílio"] not in temp:                    #Verifica se a zona de trafego ja apareceu antes
        temp.append(df.loc[line, "Zona de Tráfego do domicílio"])                   #Adicioma a zona na lista de zonas que ja apareceram
        residencias.update({df.loc[line, "Zona de Tráfego do domicílio"]
                            : [[df.loc[line, "Código do domicílio"]]]})             #Adicioma a zona como uma chave no dicionario das residencias
        domicilios.update({df.loc[line, "Zona de Tráfego do domicílio"]
                           : [df.loc[line, "Fator de expansão para domicílios"]]})  #Adicioma a zona como uma chave no dicionario dos fatores de expansão
        população.update({df.loc[line, "Zona de Tráfego do domicílio"]
                          : [df.loc[line, "Fator de expansão para domicílios"]
                             *df.loc[line, "Nro de pessoas no domicílio"]]})        #Adicioma a zona como uma chave no dicionario das populações
        automoveis.update({df.loc[line, "Zona de Tráfego do domicílio"]
                           : [df.loc[line, "Fator de expansão para domicílios"]
                              *df.loc[line, "Nro de automóveis"]]})                 #Adicioma a zona como uma chave no dicionario dos automoveis

    if df.loc[line, "Código do domicílio"] not in residencias[df.loc[line, "Zona de Tráfego do domicílio"]]:                #Verifica se residencia ja apareceu anteriomente
        residencias[df.loc[line, "Zona de Tráfego do domicílio"]].append([df.loc[line, "Código do domicílio"]])             #Adicioma os codigos dos domicilios na lista da chave da zona
        domicilios[df.loc[line, "Zona de Tráfego do domicílio"]].append(df.loc[line, "Fator de expansão para domicílios"])  #Adicioma  os fatores de expansão na lista da chave da zona
        população[df.loc[line, "Zona de Tráfego do domicílio"]].append(df.loc[line, "Fator de expansão para domicílios"]
                                                                       *df.loc[line, "Nro de pessoas no domicílio"])        #Adicioma  a população da residencia(numero de pessoas * fator de expansão) na lista da chave da zona
        automoveis[df.loc[line, "Zona de Tráfego do domicílio"]].append(df.loc[line, "Fator de expansão para domicílios"]
                                                                        *df.loc[line, "Nro de automóveis"])                 #Adicioma  o numero de automoveis da residencia(numero de automveis * fator de expansão) na lista da chave da zona
```
